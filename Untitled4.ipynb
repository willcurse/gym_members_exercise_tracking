# 3.8 What is the average height and weight of members, and is there a relationship between these variables and their workout preferences?

```python
data.columns
```

```python
avg_height=data['Height (m)'].mean()
avg_weight=data['Weight (kg)'].mean()
height_and_weight_group=data.groupby('Workout_Type').agg({'Height (m)':'mean','Weight (kg)':'mean'}).reset_index()
print(f'Average height is, {avg_height} m ')
print(f'Average weight is, {avg_weight} kg')
print(f'Group height and weight is {height_and_weight_group}')
```

```python
#height
plt.figure(figsize=(5,3))
sns.boxplot(x='Workout_Type',y='Height (m)',data=data)
plt.title('Height boxplot visual')
plt.show()

#weight
plt.figure(figsize=(5,3))
sns.boxplot(x='Workout_Type',y='Weight (kg)',data=data)
plt.title('Weight boxplot visual')
plt.show()

#relationship....between both
plt.figure(figsize=(10,6))
sns.scatterplot(x='Height (m)',y='Weight (kg)',hue='Workout_Type',data=data)
plt.title('relationship....between both height and weight')
plt.xlabel('Height')
plt.ylabel('Weight')
plt.show()
```

# 3.9 Which workout type is most popular among gym members? Does it differ between males and females?

```python
data.columns
```

```python
most_popular=data['Workout_Type'].value_counts().idxmax()
print(most_popular)

workout_gender=data.groupby(['Gender','Workout_Type']).size()
print(workout_gender)

plt.figure(figsize=(10,6))
sns.countplot(x='Workout_Type',hue='Gender',data=data)
plt.title("most popular among gym members")
plt.show()
```

#  3.10 Do members who work out more frequently burn more calories per session?

```python
group=data.groupby('Workout_Frequency (days/week)')['Calories_Burned'].mean().reset_index()
print(group)

plt.figure(figsize=(5,3))
sns.scatterplot(x='Workout_Frequency (days/week)',y='Calories_Burned',data=data)
plt.title("more frequently burn more calories per session?")
plt.xlabel('Workout_Frequency')
plt.ylabel('Calories_Burned')
plt.show()

plt.figure(figsize=(5,3))
sns.scatterplot(x='Calories_Burned',y='Workout_Frequency (days/week)',data=data)
plt.title("more frequently burn more calories per session?")
plt.xlabel('Calories_Burned')
plt.ylabel('Workout_Frequency ')
plt.show()
```
